```python
def norm_and_reshape(train_x, test_x):
    train_x_norm = train_x_array.reshape(train_x_array.shape[0], train_x_array.shape[1] * train_x_array.shape[2])
    test_x_norm = test_x_array.reshape(test_x_array.shape[0], test_x_array.shape[1] * test_x_array.shape[2])
    train_x = normalize(train_x_norm, norm='l2')
    test_x = normalize(test_x_norm, norm='l2')
    return train_x, test_x
```

```python
def reshape_x(train_x, test_x):
    train_x_reshape = train_x_array.reshape(train_x_array.shape[0], train_x_array.shape[1] * train_x_array.shape[2])
    test_x_reshape = test_x_array.reshape(test_x_array.shape[0], test_x_array.shape[1] * test_x_array.shape[2])
    return train_x_reshape, test_x_reshape
```

```python
def reshape_dimension(train_x, test_x):
    reshape_train_x = []
    reshape_test_x = []
    for x in train_x:
        reshape_train_x.append(np.asarray(x).T)
    for y in test_x:
        reshape_test_x.append(np.asarray(y).T)
    return reshape_train_x, reshape_test_x
```
